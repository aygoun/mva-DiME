```python
from datasets import load_dataset
from diffusers.pipelines.deprecated.audio_diffusion.mel import Mel
import numpy as np
import soundfile as sf
import os

# Load dataset
ds = load_dataset("teticio/audio-diffusion-breaks-256")

# Output directory
output_dir = "audio_samples"
os.makedirs(output_dir, exist_ok=True)

# SAME parameters as encoder
mel = Mel(
    x_res=256,
    y_res=256,
    hop_length=512,
    sample_rate=22050,
    n_fft=2048,
)

# Take 30 samples
subset = ds["train"].select(range(30))

for i, sample in enumerate(subset):
    image = sample["image"]  # IMPORTANT: use "image", not "audio_file"

    # Convert image back to audio
    audio = mel.image_to_audio(image)

    # Save
    output_path = os.path.join(output_dir, f"sample_{i}.wav")
    sf.write(output_path, audio, 22050)

    print(f"Saved: {output_path}")

print("Done!")
```
